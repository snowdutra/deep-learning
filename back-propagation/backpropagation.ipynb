```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

wine = load_wine()
x = wine.data
y = wine.target

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=2, shuffle=True
)

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

train_df = pd.DataFrame(x_train_scaled, columns=wine.feature_names)
train_df["target"] = y_train
test_df = pd.DataFrame(x_test_scaled, columns=wine.feature_names)
test_df["target"] = y_test

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def cross_entropy(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred + 1e-15))

def neural_network(x, weigths1, weights2):
    z1 = np.dot(x, weigths1)
    a1 = sigmoid(z1)

    z2 = np.dot(a1, weights2)
    a2 = softmax(z2)

    return a1, a2, z1, z2

def treina(x_train, y_train):
    alpha = 0.01
    num_classs = 3
    num_features = x_train.shape[1]
    hidden_size = 3


    weights1 = np.random.rand(x_train.shape[1], hidden_size) * 0.1
    weights2 = np.random.rand(hidden_size, num_classs) * 0.1

    one_hot_train = np.eye(num_classs)[y_train]

    for iteration in range(1000):
        total_error = 0

        for i in range(len(x_train)):
            x = x_train[i]
            y_true = one_hot_train[i]

            a1, a2, z1, z2 = neural_network(x, weights1, weights2)

            error = cross_entropy(a2, y_true)
            total_error += error

            delta2 = a2 - y_true
            delta1 = np.dot(weights2, delta2) * sigmoid_derivative(z1)

            grad_weights2 = np.outer(a1, delta2)
            grad_weights1 = np.outer(x, delta1)

            weights2 -= alpha * grad_weights2
            weights1 -= alpha * grad_weights1

        print(f"Iteration {iteration + 1}, Error: {total_error:.4f}")

    return weights1, weights2

weights1, weights2 = treina(x_train_scaled, y_train)
```
