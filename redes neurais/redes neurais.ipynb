```python
import numpy as np

np.random.seed(42)  # Fixar seed: resultados reproduzíveis (conceito: reprodutibilidade)

N = 1000            # Número de amostras (dataset grande)
D = 4               # Número de features (entradas)

X = np.random.randn(N, D) #data set X


true_W = np.array([[3.0], [-2.0], [1.5], [0.5]])


true_b = np.array([[1.0]])


noise = 0.5 * np.random.randn(N, 1) # y real


y = X.dot(true_W) + true_b + noise
```

```python
W = np.random.randn(D, 1)
b = np.zeros((1,1))

lr = 0.01
epochs = 2000

for epoch in range(epochs):
###Estrutura base RNN
  y_hat = X.dot(W) + b

###ERRO
  erro = y_hat - y
  loss = np.mean(erro**2)


###Gradiente
  dL_dyhat = (2/N) * erro
  dL_dW = X.T.dot(dL_dyhat)
  dL_db = np.sum(dL_dyhat, axis=0, keepdims=True)

  W = W - lr * dL_dW

  b = b - lr * dL_db

  # -------- Log --------
  if epoch % 200 == 0:
      print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")

# =========================
# 5) Resultado final
# =========================

print("\n=== Comparação final ===")
print("W aprendido:\n", W)
print("b aprendido:\n", b)

print("\nW verdadeiro:\n", true_W)
print("b verdadeiro:\n", true_b)
```

```python
# Avaliação simples: erro médio absoluto (MAE)
mae = np.mean(np.abs((X.dot(W) + b) - y))
print(f"\nMAE (erro absoluto médio): {mae:.4f}")
```
