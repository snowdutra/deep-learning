```python
import kagglehub
from pathlib import Path

dest_dir = Path("C:/Users/gustavo.telles/Desktop/deep-learning/cnn")
dest_dir.mkdir(parents=True, exist_ok=True)

dataset_handle = "zalando-research/fashionmnist"
csv_names = ["fashion-mnist_train.csv", "fashion-mnist_test.csv"]

downloaded_paths = []
for csv_name in csv_names:
    target = dest_dir / csv_name
    if target.exists():
        downloaded_paths.append(str(target))
        continue

    file_path = kagglehub.dataset_download(
        dataset_handle,
        path=csv_name,
        output_dir=str(dest_dir),
    )
    downloaded_paths.append(file_path)

print("CSV prontos em:")
for p in downloaded_paths:
    print(p)
```
